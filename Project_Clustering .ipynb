```python
from sklearn.model_selection import GridSearchCV

def silhouette_scorer(estimator, X):
    labels = estimator.fit_predict(X)
    return silhouette_score(X, labels)

param_grid = {
    'n_clusters': range(2, 11),  # Adjust the range as needed
    'affinity': ['euclidean', 'l1', 'l2', 'manhattan', 'cosine'],
    'linkage': ['ward', 'complete', 'average', 'single']
}

grid_search = GridSearchCV(
    AgglomerativeClustering(),
    param_grid,
    scoring=silhouette_scorer,  # Use the custom silhouette_scorer
    cv=5,
    n_jobs=-1
)
grid_search.fit(scaled_data)


best_params = grid_search.best_params_
best_num_clusters = best_params['n_clusters']

print("Best hyperparameters from grid search:", best_params)
print("Best number of clusters:", best_num_clusters)
```
